### **Utilizando KNN**

```python
entradas = df[[
    "raio_medio",
    "textura_media",
    "concavidade_media",
    "pontos_concavos_medios",
    "simetria_media",
    "dimensao_fractal_media"
]]
classes = df["diagnostico"]
print(f"Formato das tabelas de dados {entradas.shape} e classes {classes.shape}")
```

Treinamento - Representa 80% das amostras do conjunto de dados original,
Teste - com 20% das amostras

```python
from sklearn.model_selection import train_test_split

entradas = df[[
    "raio_medio",
    "textura_media",
    "concavidade_media",
    "pontos_concavos_medios",
    "simetria_media",
    "dimensao_fractal_media"
]]
classes = df["diagnostico"]
print(f"Formato das tabelas de dados {entradas.shape} e classes {classes.shape}")


entradas_treino, entradas_teste, classes_treino, classes_teste = train_test_split(
    entradas,
    classes,
    test_size=0.2,
    stratify=classes
)

print(f"Formato das tabelas de treino {entradas_treino.shape} e teste {classes_teste.shape}")
```

```python
# Escolha simples do melhor K via validação cruzada
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier


valores_k = list(range(1, 20))
medias = []

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for k_tmp in valores_k:
    modelo_tmp = KNeighborsClassifier(n_neighbors=k_tmp)
    scores = cross_val_score(modelo_tmp, entradas_treino, classes_treino, cv=cv, scoring='accuracy')
    medias.append(scores.mean())

best_k = valores_k[int(np.argmax(medias))]
print("Melhor K:", best_k)

plt.figure(figsize=(6,4))
plt.plot(valores_k, medias, marker='o')
plt.title("Acurácia média (CV) por K")
plt.xlabel("K")
plt.ylabel("Acurácia média (cv)")
plt.grid(True)
plt.show()
```

```python
# Importa a biblioteca
from sklearn.neighbors import KNeighborsClassifier

# Usa best_k se já foi calculado; caso contrário, mantém k=3
try:
    k = int(best_k)
except NameError:
    k = 3

modelo = KNeighborsClassifier(n_neighbors=k)

# Treina o modelo
modelo.fit(entradas_treino, classes_treino)
print("Modelo treinado com K =", k)
```

```python
# Para obter as previsões, basta chamar o método predict()
classes_encontradas = modelo.predict(entradas_teste)
print("Predição: {}".format(classes_encontradas))
```

```python
# Para determinar a quantidade de acertos (acuracia)

from sklearn.metrics import accuracy_score
acertos = accuracy_score(classes_teste, classes_encontradas)
print("Acerto médio de classificação: ", acertos)
```

```python
# 3. Realize a validação cruzada para encontrar o melhor K
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier

valores_k_novo = list(range(1, 20))
medias_novo = []

cv_novo = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for k_tmp_novo in valores_k_novo:
    modelo_tmp_novo = KNeighborsClassifier(n_neighbors=k_tmp_novo)
    scores_novo = cross_val_score(modelo_tmp_novo, entradas_treino_novo, classes_treino_novo, cv=cv_novo, scoring='accuracy')
    medias_novo.append(scores_novo.mean())

best_k_novo = valores_k_novo[int(np.argmax(medias_novo))]
print("Melhor K para o novo conjunto de features:", best_k_novo)

plt.figure(figsize=(6,4))
plt.plot(valores_k_novo, medias_novo, marker='o')
plt.title("Acurácia média (CV) por K (Novas Features)")
plt.xlabel("K")
plt.ylabel("Acurácia média (cv)")
plt.grid(True)
plt.show()
```

**Reasoning**:
Train a new KNN model using the best K found and evaluate its accuracy on the new test set. Then, generate a scatterplot to visualize the relationship between two of the new features, colored by diagnosis.



```python
# 4. Treine um novo modelo KNN
modelo_novo = KNeighborsClassifier(n_neighbors=best_k_novo)
modelo_novo.fit(entradas_treino_novo, classes_treino_novo)
print("Novo modelo treinado com K =", best_k_novo)

# 5. Avalie o modelo_novo no conjunto de teste
classes_encontradas_novo = modelo_novo.predict(entradas_teste_novo)
acertos_novo = accuracy_score(classes_teste_novo, classes_encontradas_novo)
print("Acerto médio de classificação com novas features: ", acertos_novo)

# 6. Gere um novo gráfico para visualizar as relações entre pelo menos duas das novas features
plt.figure(figsize=(8, 6))
sns.scatterplot(data=new_sub_df, x="textura_media", y="area_media", hue="diagnostico")
plt.title("Relação entre Textura Média e Área Média por Diagnóstico")
plt.xlabel("Textura Média")
plt.ylabel("Área Média")
plt.show()
```
